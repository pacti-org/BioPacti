We have the following equations

$
\begin{cases}
\dot m_{i,T} = \alpha_i \frac{N}{(p_T + \eta_i)N + c_i} \\
\dot N = - \sum_i \alpha_i \xi_i \frac{N}{(p_T + \eta_i)N + c_i} \\
\dot X_i = \beta_i A \frac{u_i m_i / \kappa_i}{1 + \sum u_j m_j / \kappa_j} \\
\dot A = - \sum_i \beta_i \xi_i A \frac{u_i m_i / \kappa_i}{1 + \sum_j u_j m_j / \kappa_j}
\end{cases}
$

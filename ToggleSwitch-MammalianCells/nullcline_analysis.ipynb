```python
# Plot nullclines for the toggle switch model with leak.
import matplotlib.pyplot as plt
import numpy as np

# Parameters
# Create sliders for these parameters with default values set as following:
alpha01 = 0.01
alpha02 = 0.01
alpha1 = 5
alpha2 = 5
delta = 1
k1 = 1
k1_bar = 5
k2 = 10
k2_bar = 10
K1 = 10
K2 = 20

# Slider ranges
alpha01_range = (0, 10, 0.01)
alpha02_range = (0, 10, 0.01)
alpha1_range = (0, 100, 0.1)
alpha2_range = (0, 100, 0.1)
delta_range = (0, 20, 0.1)
k1_range = (0, 100, 0.1)
k2_range = (0, 100, 0.1)
k1_bar_range = (0, 100, 0.1)
k2_bar_range = (0, 100, 0.1)
K1_range = (0, 100, 0.1)
K2_range = (0, 100, 0.1)

# Now create sliders with nullclines:
from ipywidgets import interact, FloatSlider
@interact(alpha01=FloatSlider(min=alpha01_range[0], max=alpha01_range[1], step=alpha01_range[2], value=alpha01),
          alpha02=FloatSlider(min=alpha02_range[0], max=alpha02_range[1], step=alpha02_range[2], value=alpha02),
          alpha1=FloatSlider(min=alpha1_range[0], max=alpha1_range[1], step=alpha1_range[2], value=alpha1),
          alpha2=FloatSlider(min=alpha2_range[0], max=alpha2_range[1], step=alpha2_range[2], value=alpha2),
          delta=FloatSlider(min=delta_range[0], max=delta_range[1], step=delta_range[2], value=delta),
          k1=FloatSlider(min=k1_range[0], max=k1_range[1], step=k1_range[2], value=k1),
          k2=FloatSlider(min=k2_range[0], max=k2_range[1], step=k2_range[2], value=k2),
          k1_bar=FloatSlider(min=k1_bar_range[0], max=k1_bar_range[1], step=k1_bar_range[2], value=k1_bar),
          k2_bar=FloatSlider(min=k2_bar_range[0], max=k2_bar_range[1], step=k2_bar_range[2], value=k2_bar),
          K1=FloatSlider(min=K1_range[0], max=K1_range[1], step=K1_range[2], value=K1),
          K2=FloatSlider(min=k2_range[0], max=k2_range[1], step=K2_range[2], value=K2))

def plot_nullclines(alpha01, alpha02, alpha1, alpha2, delta, k1, k2, k1_bar, k2_bar, K1, K2):
    # Nullclines
    x1 = np.linspace(0, 200, 100)
    x2 = np.linspace(0, 200, 100)
    X1, X2 = np.meshgrid(x1, x2)
    U = (K1 * (alpha01 + alpha1*X1)) / (1 + k1*X1 + k2*X2) - delta * X1
    V = (K2 * (alpha02 + alpha2*X2)) / (1 + k1_bar*X1 + k2_bar*X2) - delta * X2
    import matplotlib.lines as mlines 

    # Plot nullclines
    fig, ax = plt.subplots()

    # Label the contours
    ax.contour(X1, X2, U, levels=[0], colors='r')
    ax.contour(X1, X2, V, levels=[0], colors='b')

    # Create proxy artists
    red_line = mlines.Line2D([], [], color='r', label='dm_i/dt=0')
    blue_line = mlines.Line2D([], [], color='b', label='dm_j/dt=0')

    # Add the legend manually to the current Axes.
    ax.legend(handles=[red_line, blue_line], loc='upper right')

    ax.set_xlabel('m_i')
    ax.set_ylabel('m_j')
    ax.set_title('Nullclines')
    # Set xlim and ylim from -10 to 100
    ax.set_xlim(-10, 80)
    ax.set_ylim(-10, 120)
    # plt.show()

# Call the function once to display the plot
plot_nullclines(alpha01, alpha02, alpha1, alpha2, delta, k1, k2, k1_bar, k2_bar, K1, K2)
```
